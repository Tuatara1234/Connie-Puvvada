I also found out the standard deviation for:
    - Men's height -  2.9 inches
    - Women's height - 2.7 inches
       
I also know the Central Limit Theorem. It states:

 - the sampling distribution is normally distributed 
 - the sampling distribution mean equals to the the population mean.  

From normal distribution theory, I know:
 - one standard deviation covers 68.2% of the sample population
 - two standard deviations cover 95.6% of the sample population
 - three standard deviations cover 99.9% of the sample population
 
From the information above, I can visualize & calculate the sampling distribution of men and women's heights. 
 

## Visualize the Men and Women Height Normal Distribution

```python
# First, import Python packages

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = (8, 4)
```

#### 1.  Assign men and women mean and standard deviation values

```python
# Create men and women mean height and standard deviation variables.

men_mu = 69.2            # men's mean height in the US
men_sigma = 2.9          # men's height standard deviation 

women_mu = 63.7          # women's mean height in the US
women_sigma = 2.7        # women's height standard deviation
```

#### 2. Generate a normal distribution with sample size 100000

```python
# Generate a sample normal distribution with size one hundred thousand for men and women

men_sample = np.random.normal(men_mu, men_sigma, 100000)

women_sample = np.random.normal(women_mu, women_sigma, 100000)
```

#### 3. Verify the mean and standard deviation for men and women

```python
# Verify the mean for men
print ("Men's height normal distribution mean is: "), np.mean(men_sample)
print ("Men's height normal distribution standard deviation is: "), np.std(men_sample)
```

```python
# Verify the mean for women
print ("Women's height normal distribution mean is: "), np.mean(women_sample)
print ("Women's height normal distribution standard deviation is: "), np.std(women_sample)
```

##### As you can see, the random sampling distribution mean for one hundred thousand men and women is very close to the population mean. 

#### 4. Plot the men and women height distribution charts

```python
# Plot men's & women distribution

sns.set_style('dark', {'axes.facecolor': 'gray'})
sns.distplot(men_sample, bins = 20, hist_kws = {'edgecolor': 'black'}, \
            color = 'blue')
sns.distplot(women_sample, bins = 20, hist_kws = {'edgecolor': 'black'}, \
            color = 'pink')
plt.title('Sampling Height Distribution for Men and Women in the U.S.', fontsize = 16)
plt.xlabel('Height in inches', fontsize = 14)
plt.ylabel('Probability', fontsize = 14)
plt.xticks(np.arange(55, 90, 5))
plt.show()
```

## Calculate the minimum height of 2.2% of tallest men & women in inches.

For normal distribution, We know 2 positive and negetive standard deviations are 95.6%. 

```python
print('Therefore, the shortest and the tallest make up:'), str((100 - 95.6)) + str('%')
```

To find out the tallest 2.2%, we add the 2 positive standard deviations from the mean of each group.

```python
# We know 2 standard deviations is 95.6% of the normal distribution 
print("Men's mean height is: "), men_mu
print("Men's standard deviation is: "), men_sigma
print("Minimum men's height for the top 2.2% is: "), (men_mu + (2*men_sigma))
```

```python
# We know 2 standard deviations is 95.6% of the normal distribution 
print("Women's mean height: "), women_mu
print("Wen's standard deviation is:"), women_sigma
print("Minimum women's height for the top 2.2% is: "), (women_mu + (2*women_sigma))
```

##### Here you go! With the simple yet powerful Central Limit Theorem and Normal Distribution theory, we can get the top 2.2% minium height for men and women in the U.S.
